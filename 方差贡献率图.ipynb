```python
import pandas as pd  
import numpy as np  
from scipy import stats  
import matplotlib.pyplot as plt  
  
# 假设你已经有了一个包含偏度和峰度结果的DataFrame  
# 这里我们重新使用上面的代码片段来生成这个DataFrame  
  
# 假设df是你的原始DataFrame  
np.random.seed(0)  # 为了可复现性设置随机种子  
data = {  
    'col1': np.random.randn(1000),  
    'col2': np.random.normal(loc=2, scale=1.5, size=1000),  
    'col3': np.random.chisquare(df=3, size=1000)  # 自由度为3的卡方分布  
}  
df = pd.DataFrame(data)  
  
# 定义函数并计算偏度和峰度  
def calculate_skew_kurt(series):  
    skew = stats.skew(series)  
    kurt = stats.kurtosis(series) + 3  
    return pd.Series([skew, kurt])  
  
skew_kurt_results = df.apply(calculate_skew_kurt) 
skew_kurt_results=skew_kurt_results.T
skew_kurt_results.columns = ['Skewness', 'Kurtosis']  
  
# 绘制偏度折线图  
plt.subplot(2, 1, 1)  
plt.plot(df.columns, skew_kurt_results['Skewness'], marker='o')  
plt.title('Skewness of Samples')  
plt.xlabel('Sample')  
plt.ylabel('Skewness')  
  
# 绘制峰度折线图  
plt.subplot(2, 1, 2)  
plt.plot(df.columns, skew_kurt_results['Kurtosis'], marker='o')  
plt.title('Kurtosis of Samples')  
plt.xlabel('Sample')  
plt.ylabel('Kurtosis')  
  
# 展示图形  
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd  
from sklearn.decomposition import PCA  
import matplotlib.pyplot as plt  
relative_abundance=pd.read_csv("./relative_abundance.csv",index_col=0)
# 假设relative_abundances是一个DataFrame，其中行是OTU，列是样本  
# 这里我们使用一个随机生成的DataFrame作为示例  
import numpy as np  
#np.random.seed(0)  
#num_otus = 50  # 假设有50个OTU  
#num_samples = 100  # 假设有100个样本  
#relative_abundances = pd.DataFrame(np.random.rand(num_otus, num_samples), index=[f'OTU_{i}' for i in range(num_otus)], columns=[f'Sample_{i}' for i in range(num_samples)])  
  
# 进行PCA分析，假设我们保留10个主成分以计算方差贡献率  
pca = PCA(n_components=10).fit(relative_abundance)    
  
# 获取每个主成分的方差贡献率  
explained_variance_ratio = pca.explained_variance_ratio_  
  
# 绘制方差贡献率图  
plt.figure(figsize=(10, 6))  
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5, align='center')  
plt.ylabel('Explained variance ratio')  
plt.xlabel('Principal component index')  
plt.title('Explained Variance Ratio by Principal Components')  
plt.xticks(range(1, len(explained_variance_ratio) + 1))  
plt.show()
# 绘制碎石图  
plt.figure(figsize=(10, 6))  
plt.plot(explained_variance_ratio.cumsum(), marker='o')  
plt.xlabel('Number of Components')  
plt.ylabel('Cumulative Explained Variance')  
plt.title('Scree Plot')  
plt.grid(True)  
plt.show()
```

```python
clr_dff = pd.read_csv("./clr_dff.csv",index_col=0)
import pandas as pd    
from sklearn.decomposition import PCA    
import matplotlib.pyplot as plt       
import numpy as np    
  
pca = PCA(n_components=10).fit(clr_dff)      
  
# 获取每个主成分的方差贡献率  
explained_variance_ratio = pca.explained_variance_ratio_    
  
# 绘制方差贡献率图  
plt.figure(figsize=(10, 6))  
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5, align='center')  
  
# 在每个条形图的上方添加文本  
for i, evr in enumerate(explained_variance_ratio):  
    plt.text(i + 1, evr, '{:.2%}'.format(evr), ha='center', va='bottom')  # 这里的'{:.2%}'是为了将浮点数格式化为百分比  
  
plt.ylabel('Explained variance ratio')  
plt.xlabel('Principal component index')  
plt.title('Explained Variance Ratio by Principal Components')  
plt.xticks(range(1, len(explained_variance_ratio) + 1))  
plt.show()
# 绘制碎石图  
plt.figure(figsize=(10, 6))  
plt.plot(explained_variance_ratio.cumsum(), marker='o')  
plt.xlabel('Number of Components')  
plt.ylabel('Cumulative Explained Variance')  
plt.title('Scree Plot')  
plt.grid(True)  
plt.show()
```
